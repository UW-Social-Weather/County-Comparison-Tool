```python
import pandas as pd
```

```python
county_US = pd.read_csv("SVI2018_US_COUNTY.csv")
```

```python
county_US[['ST',"STATE","ST_ABBR","COUNTY","LOCATION", 'SPL_THEME1','SPL_THEME2','SPL_THEME3','SPL_THEME4','SPL_THEMES','RPL_THEMES']]
```

```python
county_US.loc[:,'SPL_THEMES']
```

```python
county_US['SVI Group'] = pd.qcut(county_US['SPL_THEMES'], q=[0, 0.25, 0.5, 0.75, 1], labels=['low','mid-low','mid-high','high'])
```

```python
county_US.to_csv("US_County_SVI_Group.csv", index=False)
```
